RNNs face the same issues as ANNs in terms of vanishing gradient since gradV is dependent on every Ws change to the corresponding hidden layer and from U with the same logic which means if these are all small will lead to vanishing and if these are all large will lead to exploding gradient.


As a result RNN does not perform well for long sentences when considering things like a text predictor.
to mitigate we can again use gradient clipping to solve exploding gradient but to solve vanishing gradient we can use LSTM or GRU

# LSTM

Essentially an advanced RNN with 3 sections, the 3 gates:

-Forget gate where irrelevant information is forgot (basically our dropout)
-Input gate is where we ad/update new information e.g. stock price at this time
-Output gate is where the information is then sent on to the next cell.

Just like in RNN we have the hidden states H_t-1 is the hidden state from the previous time step and H_t is the current step hidden state.



In addition we also have a Cell state in LSTM C_t-1 and 
C_t which carries information along all the previous
time states.
so C is like the long term memory and H is the short term memory which is all 
we had before

Forget gate:
f_t = sigma(x_t*U_t + H_t-1 * W_f)
(x_t is the input from the current input
U_f is the weight metric associated with that input

H_t-1 is the value of the previous hidden layer
and W_f is the weight metric associated with that previous hidden state)

and this is the sigmoid function so f_t will be a number between 1 and 0.

this f_t is multiplied by the input Cell state (C_t-1)
and so if f_t = 0, our C_t-1 * f_t = 0 and so we will forget everything from the past.
if f_t = 1 then our C_t-1 will remain so forget nothing


e.g. idealy we would start forgetting things from previous sentences i.e. when we reach full stops
